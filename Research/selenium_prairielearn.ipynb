```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd

username = "ENTER"
password = "ENTER"

service = Service(executable_path = "./chromedriver")
driver = webdriver.Chrome(service = service)
driver.get("https://us.prairielearn.com/pl/login")

#Selects UIUC
schoolLink = driver.find_element(By.PARTIAL_LINK_TEXT, "University of Illinois Urbana-Champaign (UIUC)")
schoolLink.click()
time.sleep(2)

#Passes login and pass keys
driver.find_element(By.XPATH, "//*[@id=\"i0116\"]").send_keys(username)
next = driver.find_element(By.XPATH, "//*[@id=\"idSIButton9\"]")
next.click()
time.sleep(2)
driver.find_element(By.XPATH, "//*[@id=\"i0118\"]").send_keys(password)
passnext = driver.find_element(By.XPATH, "//*[@id=\"idSIButton9\"]")
passnext.click()
time.sleep(2)



#Pulls html file from PL
pageSource = driver.page_source
fileToWrite = open("page_source.html")
soup = BeautifulSoup(fileToWrite, "html.parser")

a_tags = soup.find_all("a", href = lambda href: href and href.startswith("/pl/course_instance"))
for tag in a_tags:
    #Get Class Names
    classname = tag.text.strip()
    print(f"Class Name = ", classname)
    clickIntoClassName = driver.find_element(By.PARTIAL_LINK_TEXT, classname)
    clickIntoClassName.click()
    time.sleep(2)
    
    #Pulls Current Class HTML 
    classHTML = driver.page_source
    newSoup = BeautifulSoup(classHTML, "html.parser")
    
    #Pulls Class Assignments
    class_assignment_a_tags = newSoup.find_all("a", href = lambda href: href and href.startswith("/pl/course_instance/145219/assessment"))
    for tags in class_assignment_a_tags:
        assignment = tags.text.strip()
        print(f"    Assignment = ", assignment)
        
    time.sleep(3)  
    #Loops back to main PL
    exitToHome = driver.find_element(By.CLASS_NAME, "navbar-brand")
    exitToHome.click()
    
```
